```python
# Importar las bibliotecas
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns


# Generar datos de ejemplo
np.random.seed(42)
data = {
    'edad': np.random.randint(18, 65, 100),
    'ingreso': np.random.normal(35000, 15000, 100),
    'gasto': np.random.normal(20000, 10000, 100),
    'satisfaccion': np.random.randint(1, 6, 100)
}


# Crear DataFrame
df = pd.DataFrame(data)


# Mostrar primeras filas
print("Primeras 5 filas:")

display(df.head())

# Estad√≠sticas descriptivas
print("\nEstad√≠sticas descriptivas:")
display(df.describe())

# Visualizaci√≥n
plt.figure(figsize=(10, 6))
sns.scatterplot(x='ingreso', y='gasto', hue='satisfaccion', 
                palette='viridis', data=df)
plt.title('Relaci√≥n entre Ingreso y Gasto')
plt.show()

# Simple modelo de clustering
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler


# Normalizar datos
scaler = StandardScaler()
scaled_data = scaler.fit_transform(df[['ingreso', 'gasto']])


# Aplicar K-Means
kmeans = KMeans(n_clusters=3, random_state=42)
df['cluster'] = kmeans.fit_predict(scaled_data)


# Visualizar clusters
plt.figure(figsize=(10, 6))
sns.scatterplot(x='ingreso', y='gasto', hue='cluster', 
                palette='Set1', data=df)
plt.title('Segmentaci√≥n de clientes con K-Means')
plt.show()
print("¬°El notebook funciona correctamente!")
```
